```python
from fastai.vision.all import *
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('./fairface_label_train.csv')

df_val = pd.read_csv('./fairface_label_val.csv')
```

```python
print("Training Data:")

print(df_train.head())

print('\n', 'Dataset shape' , df_train.shape)

print('\n', 'Total training images:', df_train.shape[0])
```

```python
print('\n', 'Validation Data:')

print(df_val.head())

print('\n', 'Dataset shape' , df_val.shape)

print('\n', 'Total validation images:', df_val.shape[0])
```

```python
print("Basic Information:")
print(df_train.nunique())
```

```python
print("\nAge Distribution:")
print(df_train['age'].value_counts())
```

```python
print("\nGender Distribution:")
print(df_train['gender'].value_counts())
```

```python
print("\nRace Distribution:")
print(df_train['race'].value_counts())
```

```python
print("\nService Test by Age Distribution:")
print(df_train.groupby('age')['service_test'].mean())
```

```python
sample_train_df = df_train.sample(5)

plt.figure(figsize=(15, 10))
subplot_index = 1

for _, row in sample_train_df.iterrows():
    img_path = './Dataset/' + row['file'] 
    
    img = Image.open(img_path)
    
    plt.subplot(1, 5, subplot_index)
    plt.imshow(img)
    plt.axis('off')
    
    plt.title(f"{row['age']}\n{row['gender']}\n{row['race']}")
    
    subplot_index += 1 

plt.show()
```
